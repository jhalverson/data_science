### Trimmed mean

```python
from scipy import stats
from scipy import random
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('halverson')
```

```python
x = random.randn(100)
```

```python
plt.hist(x)
```

```python
stats.trim_mean(x, 0.1), x.mean()
```

### Hexagonal binning plot

```python
import numpy as np
from numpy.random import multivariate_normal
```

```python
mean = [0, 2]
cov = [[1, 0], [0, 5]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T
```

```python
plt.hexbin(x, y, gridsize=50, cmap='inferno')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
```

### Violin plot

```python
pos = range(1, 6)
data = [np.random.normal(loc=0, scale=1, size=25) for _ in pos]
```

```python
plt.violinplot(data, pos, showmeans=False, showextrema=True, showmedians=True, bw_method=0.5)
```
