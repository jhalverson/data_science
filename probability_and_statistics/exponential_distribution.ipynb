# Exponential distribution

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('halverson')
```

```python
from scipy.stats import expon
x = np.linspace(0, 5, num=250)
rate = 2.0
inverse_lambda = 1.0 / rate
plt.plot(x, expon.pdf(x, scale=1/0.5), 'k-', label=r'$\lambda=1/2$')
plt.plot(x, expon.pdf(x, scale=1/1.0), 'r-', label=r'$\lambda=1$')
plt.plot(x, expon.pdf(x, scale=1/1.5), 'b-', label=r'$\lambda=3/2$')
plt.xlabel(r'$x$', fontsize=22)
plt.ylabel(r'$P(x)=\lambda e^{-\lambda x}$', fontsize=22)
plt.legend(fontsize=18)
```

### The exponential distribution describes the amount of time between arrivials in a Poisson process.

### The exponential distribution is memoryless in the sense $P(X>s+t|X>s)=P(X>t)$, where s and t are positive constants. It is the continuous analogue of the geometric distribution.

### Imagine you see a shooting star every 15 minutes on average or 4 per hour so lambda = 4. Let's draw some variants:

```python
waiting_times = expon.rvs(scale=1/4.0, size=1000)
plt.hist(waiting_times, bins=20)
plt.xlabel(r'$t$')
plt.ylabel('Count')
```

### The mean waiting time is

```python
waiting_times.mean()
```

### This implies four events per time window as expected.
