# Transfer learning with a CNN

The purpose of this notebook is to illustrate transfer learning by retraining the output layer. We will use the flowers data set which has four classes and hundreds of images.

```python
import numpy as np
import tensorflow as tf
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
import os
import glob
```

```python
dsy = glob.glob(os.path.join('datasets', 'flower_photos', 'daisy') + "/*.jpg")
```

```python
import skimage
```

```python
dsy[0]
```

```python
img = mpimg.imread(dsy[0])
plt.imshow(img)
```

```python
import skimage.transform

img_resized = skimage.transform.resize(img, output_shape=(299, 299))
plt.imshow(img_resized)
```

```python
import skimage.util
```

```python
img_cropped = skimage.util.crop(img, crop_width=((100, 100), (50, 50), (0, 0)))
plt.imshow(img_cropped)
```
