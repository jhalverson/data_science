###A normal probability plot is used to test whether data follows a certain distribution: 

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

###Generate samples from a normal distribution and make a histogram: 

```python
dist_mean = 80.0
dist_std = 12.5
samples = norm.rvs(loc=dist_mean, scale=dist_std, size=400)
```

```python
n, bins, patches = plt.hist(samples, bins=20)
```

###Make a normal probability plot

```python
from scipy.stats import probplot
fig = plt.figure()
ax = fig.add_subplot(111)
res=probplot(samples, dist="norm", plot=ax)
```

###The plot shows that indeed the data are Gaussian with only slight departures at the tails as expected for a distribution of this size.
