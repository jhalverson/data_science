# Kernel PCA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('halverson')
```

PCA and LDA are linear transformation techniques. For nonlinear problems these techniques will most likely be insufficient. The kernel PCA method will allow us to transform data that is not linearly separable onto a new, lower-dimensional subspace that is suitable for linear classifiers like logistic regression and linear SVM.

The essense of the method is to transform the data to a higher-dimensional space, where the data are separable, by using a nonlinear combination of the original features. This is similar to kernel SVM. Once in the higher-dimensional space, PCA is used to project the data onto a lower-dimensional space where it is linearly separable.

Different kernel functions are available. Note that this is a computational expensive method because it creates a n x n matrix of all pairs of samples evalated by the kernel function.

### Example 1

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=100)
plt.plot(X[y==0, 0], X[y==0, 1], 'wo')
plt.plot(X[y==1, 0], X[y==1, 1], 'ro')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
```

```python
from sklearn.decomposition import KernelPCA
kpca = KernelPCA(n_components=2, kernel='rbf', gamma=15.0)
X_kpca = kpca.fit_transform(X)
```

```python
plt.plot(X_kpca[y==0, 0], X_kpca[y==0, 1], 'wo')
plt.plot(X_kpca[y==1, 0], X_kpca[y==1, 1], 'ro')
plt.xlabel('PC 1')
plt.ylabel('PC 2')
```

### Example 2

```python
from sklearn.datasets import make_circles
X, y = make_circles(n_samples=1000, noise=0.1, factor=0.2)
plt.plot(X[y==0, 0], X[y==0, 1], 'wo')
plt.plot(X[y==1, 0], X[y==1, 1], 'ro')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
```

```python
kpca = KernelPCA(n_components=2, kernel='rbf', gamma=15.0)
X_kpca = kpca.fit_transform(X)
```

```python
plt.plot(X_kpca[y==0, 0], X_kpca[y==0, 1], 'wo')
plt.plot(X_kpca[y==1, 0], X_kpca[y==1, 1], 'ro')
plt.xlabel('PC 1')
plt.ylabel('PC 2')
```

The data are now linearly separable. Note that the value a gamma needs to be tuned to achieve this separation.
